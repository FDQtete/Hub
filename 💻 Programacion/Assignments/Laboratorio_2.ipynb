```python
num_uno = int(input("Ingrese el primer numero: "))
num_dos = int(input("Ingrese el segundo numero: "))
numero_mayor = 0
numero_menor = 0

if (num_uno > num_dos):
  numero_mayor = num_uno
  numero_menor = num_dos
elif num_dos > num_uno:
  numero_mayor = num_dos
  numero_menor = num_uno
elif num_dos == num_uno:
    print("No se pudo generar cuenta regresiva");

while numero_mayor >= numero_menor:
  print(numero_mayor)
  numero_mayor-=1
```
